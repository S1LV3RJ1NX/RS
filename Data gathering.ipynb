## Getting user reviews per restaurant

```python
reviews = defaultdict()
```

```python
url = "https://developers.zomato.com/api/v2.1/reviews?res_id=18871246"
resp = requests.get(url,headers=headers1)
json_data = json.loads(resp.text)
```

```python
print(json_data)
```

```python
for res_id in df.res_id:
    url = "https://developers.zomato.com/api/v2.1/reviews?res_id="+str(res_id)
    resp = requests.get(url,headers=headers1)
    json_data = json.loads(resp.text)
    reviews[res_id] = json_data['user_reviews']
    
    print(res_id)
    
    
```

```python
final_review = defaultdict(list)
for key in reviews.keys():
    for review in reviews[key]:
        final_review['res_id'].append(key)
        final_review['user_id'].append(review['review']['id'])
        final_review['rating'].append(review['review']['rating'])
        final_review['rating_text'].append(review['review']['review_text'])
    
```

```python
raw_df_rew = pd.DataFrame.from_dict(final_review)
```
