```python
# Preparing data table for analysis
user_rating= pd.pivot_table(data=reviews, values='rating', index='user_id', columns='res_id')
user_rating.head()
```

```python
sample_hotel = user_rating[50474]
```

```python
hotels_like_sample_hotel = user_rating.corrwith(sample_hotel)
corr_sample_hotel = pd.DataFrame(hotels_like_sample_hotel, columns=['Correlation'])  
corr_sample_hotel.dropna(inplace=True)
corr_sample_hotel.head()
```
